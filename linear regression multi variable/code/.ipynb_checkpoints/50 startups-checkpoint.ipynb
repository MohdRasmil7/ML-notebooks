```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,RobustScaler
```

```python
data=pd.read_csv('50_Startups.csv')
```

```python
data.sample(10)
```

```python
data.describe()
```

# getting dummies for state column

```python
dummies=pd.get_dummies(data['State'])
dummies
```

## concating the dummies

```python
merged_data=pd.concat([data,dummies],axis=1)
merged_data.sample(3)
```

```python
final_data=merged_data.drop(['State','California'],axis=1)
```

## Visualization

```python
categorical_values=[]
for i in final_data.columns:
    if(final_data[i].nunique()<15):
        categorical_values.append(i)
categorical_values
```

```python
numerical_values=[]
for i in final_data.columns:
    if(final_data[i].nunique()>15):
        numerical_values.append(i)
numerical_values
```

```python
# function takes a single dataframe column name as argument and
# plots a 1 x 2 subplot contains (pie plot and bar plot)
def plot_categorial(col_name):
    fig, ax = plt.subplots(
        nrows=1, ncols=2, figsize=(14, 5)
    )  # sets 1 x 2 subplots with figure size of (14,5)
    final_data[col_name].value_counts().plot.pie(
        
        labels=final_data[col_name].unique(),
        shadow=True,
        startangle=180,
        ax=ax[0],
    )  # plots pie chart of given column with % percentage

    final_data[col_name].value_counts().plot.bar(
        ax=ax[1]
    )

    fig.suptitle(col_name.title())  # set plot's title
    plt.show()  # shows the plot
```

```python
# function takes a single dataframe column name as argument and
# plots a single plot contains kernel density estimate plot
def plot_kde(col_name):
    facet = sns.FacetGrid(
        final_data, aspect=4, palette="bright"
    )  # sets a seaborn FacetGrid with hue = "Outcome" and palette as "bright"
    facet.map(
        sns.kdeplot, col_name, shade=True
    )  # kernel density plot of the given dataframe column 
    facet.set(
        xlim=(0, final_data[col_name].max())
    )  # sets the x axis from 0 to maximum value of the column
    facet.add_legend()  # add legend to the plot
    plt.title(
        col_name +" distribution"
    )  # sets the title for the plot as "<column name> respect to Outcome"
    plt.show()  # shows the plot
```

```python
for i in categorical_values:
    plot_categorial(i)
```

```python
for i in numerical_values:
    plot_kde(i)
```

## creating model

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_model=LinearRegression()
```

## Train test split

```python
features=final_data.drop('Profit',axis=1)
```

```python
features.sample(3)
```

```python
Target=final_data['Profit']
```

```python
Target.sample(4)
```

```python
X_train, X_test, y_train, y_test =train_test_split(features,Target,train_size=0.80,random_state=100)
```

```python
print(X_train.shape)
```

```python
print(X_test.shape)
```

## Feature scaling

```python
sc=RobustScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

## fitting the model

```python
lin_model.fit(X_train,y_train)
```

```python
lin_model.predict(X_test)
```

```python
lin_model.score(X_test,y_test)
```
