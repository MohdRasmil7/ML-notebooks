```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

X=data[[ 'Age_08_04', 'Mfg_Month', 'Mfg_Year', 'KM', 'HP', 'Met_Color', 'Automatic', 'cc', 'Doors',
       'Cylinders', 'Gears', 'Quarterly_Tax', 'Weight', 'Mfr_Guarantee',
       'BOVAG_Guarantee', 'Guarantee_Period', 'ABS', 'Airbag_1', 'Airbag_2',
       'Airco', 'Automatic_airco', 'Boardcomputer', 'CD_Player',
       'Central_Lock', 'Powered_Windows', 'Power_Steering', 'Radio',
       'Mistlamps', 'Sport_Model', 'Backseat_Divider', 'Metallic_Rim',
       'Radio_cassette', 'Tow_Bar']]

vif_data=pd.DataFrame()
vif_data['feature']=X.columns

vif_data['vif']=[variance_inflation_factor(X.values,i)  for i in range(len(X.columns))]

vif_data
```

## remove unneccesary columns

```python
data_cleaned=data[['Price','Age_08_04','KM','HP','cc','Doors','Gears','Quarterly_Tax','Weight']]
```

```python
data_cleaned
```

### finding duplicates and removing

```python
data_cleaned[data_cleaned.duplicated()]
```

```python
data_cleaned2=data_cleaned.drop_duplicates()
```

```python
data_cleaned2.duplicated().any()
```

## finding null values and treating them

```python
data_cleaned2.isnull().sum()
```

##  doing train test split

```python
features=data_cleaned2.drop('Price',axis=1)
features.head()
```

```python
target=data_cleaned2['Price']
target.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test =train_test_split(features,target,train_size=0.80)
```

```python
print(X_train.shape)
```

```python
print(y_test.shape)
```

## feature scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

## creating linear model

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_model=LinearRegression()
```

```python
lin_model.fit(X_train,y_train)
```

```python
y_pred=lin_model.predict(X_test)
y_pred
```

# checking accuracy

```python
lin_model.score(X_test,y_test)
```
