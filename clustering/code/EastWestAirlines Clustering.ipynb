## Hirearchical clustering

```python
df1=pd.read_excel('EastWestAirlines.xlsx',sheet_name="data")
df1.head()
```

```python
df1.info()
```

```python
df1.duplicated().any()
```

```python
for x in df1.columns:
     sns.distplot(df1[x])
     plt.show()
```

```python
scaler=MinMaxScaler()
scaled_df1 = pd.DataFrame(scaler.fit_transform(df1.iloc[:,1:]))
scaled_df1
```

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as shc
```

```python
silhouette_scores = [] 

for n_cluster in range(2, 8):
    silhouette_scores.append( 
        silhouette_score(scaled_df1, AgglomerativeClustering(n_clusters = n_cluster).fit_predict(scaled_df1))) 
    
# Plotting a bar graph to compare the results 
k = [2, 3, 4, 5, 6,7] 
plt.bar(k, silhouette_scores) 
plt.xlabel('Number of clusters', fontsize = 10) 
plt.ylabel('Silhouette Score', fontsize = 10) 
plt.show() 
```

## KMEANS

```python
from sklearn.cluster import KMeans
```

```python
df2=pd.read_excel('EastWestAirlines.xlsx',sheet_name='data')
```

```python
df2.head()
```

```python
df2.info()
```

```python
df2.describe()
```

```python
sc=StandardScaler()
X_scaled=sc.fit_transform(df2.iloc[:,1:])
```

```python
X_scaled
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.scatter(range(1, 11), wcss,color='red')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
km=KMeans(n_clusters=6)
km.fit(X_scaled)
```

```python
km.labels_
```

```python
df2['clusters_kmeans']=km.labels_
df2.head()
```

```python
silhouette_scores = [] 

for n_cluster in range(2, 8):
    silhouette_scores.append( 
        silhouette_score(X_scaled, KMeans(n_clusters = n_cluster).fit_predict(X_scaled))) 
    
# Plotting a bar graph to compare the results 
k = [2, 3, 4, 5, 6,7] 
plt.bar(k, silhouette_scores) 
plt.xlabel('Number of clusters', fontsize = 10) 
plt.ylabel('Silhouette Score', fontsize = 10) 
plt.show() 
```
