```python
import seaborn as sns
for i in df1.columns:
    if df1[i].dtypes!='object':
        print(i,"\n")
        sns.distplot(df1[i])
        plt.show()
```

```python
sns.scatterplot(df1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X_scaled=sc.fit_transform(df1.iloc[:,1:])
```

```python
X_scaled
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3)
km.fit_predict(X_scaled)
```

```python
import numpy as np
np.unique(km.labels_)
```

```python
df1['cluster']=km.labels_
```

```python
df1.head()
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.scatter(range(1, 11), wcss,color='red')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
silhouette_scores = [] 

for n_cluster in range(2, 8):
    silhouette_scores.append( 
        silhouette_score(X_scaled, KMeans(n_clusters = n_cluster).fit_predict(X_scaled))) 
    
# Plotting a bar graph to compare the results 
k = [2, 3, 4, 5, 6,7] 
plt.bar(k, silhouette_scores) 
plt.xlabel('Number of clusters', fontsize = 10) 
plt.ylabel('Silhouette Score', fontsize = 10) 
plt.show() 
```

since we got maximum score at cluster 2 and 4 . so we take 4  as cluster value 

```python
clusters = KMeans(n_clusters=4)
clusters.fit(X_scaled)
```

```python
clusters.labels_
```

```python
df1['clusters_kmeans']=clusters.labels_
```

```python
df1.head()
```

## Hirearchical clusering

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import MinMaxScaler
import scipy.cluster.hierarchy as shc
from sklearn.metrics import silhouette_score
```

```python
df2=pd.read_csv('crime_data.csv')
```

```python
trans = MinMaxScaler()
data = pd.DataFrame(trans.fit_transform(df2.iloc[:,1:]))
data
```

```python
plt.figure(figsize =(10, 6)) 
plt.title('Visualising the data') 
Dendrogram = shc.dendrogram((shc.linkage(data, method ='ward'))) 
```

```python
silhouette_scores = [] 

for n_cluster in range(2, 8):
    silhouette_scores.append( 
        silhouette_score(data, AgglomerativeClustering(n_clusters = n_cluster).fit_predict(data))) 
    
# Plotting a bar graph to compare the results 
k = [2, 3, 4, 5, 6,7] 
plt.bar(k, silhouette_scores) 
plt.xlabel('Number of clusters', fontsize = 10) 
plt.ylabel('Silhouette Score', fontsize = 10) 
plt.show() 
```

```python
agg=AgglomerativeClustering(n_clusters=2)
agg.fit(data)
```

```python
agg.labels_
```

```python
df2['cluster_agg']=agg.labels_
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
df3=pd.read_csv('crime_data.csv')
```

```python
sc=StandardScaler()
data_scaled=sc.fit_transform(df3.iloc[:,1:])
```

```python
data_scaled
```

```python
db=DBSCAN(eps=1.1,min_samples=3)
```

```python
db.fit(data_scaled)
```

```python
db.labels_
```

```python
df3['db_clusters']=db.labels_
```
